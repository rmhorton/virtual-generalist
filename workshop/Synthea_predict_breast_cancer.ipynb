```
%sql
with
latest_patient_start as (
  select patient, max(date(start)) latest_start from encounters 
    group by patient
),
most_recent_encounters as (
  select encounters.* 
    from encounters inner join latest_patient_start 
      on encounters.patient = latest_patient_start.patient 
      and date(encounters.start) = latest_patient_start.latest_start
    where date(encounters.start) > date('2021-01-01')
)
select e.id encounter, p.first, p.last, floor(datediff(date(e.start), date(p.birthdate))/365.24) age, p.race, p.ethnicity, p.gender
    , c.description
  from patients p
  inner join most_recent_encounters e on p.id = e.patient
    left join conditions c on c.patient = e.patient and date(c.start) <= date(e.stop)  where c.description = 'Malignant neoplasm of breast (disorder)'

```

# Breast Cancer

Predict whether breast cancer will be diagnosed in a given encounter. Exclude patients who have a current diagnosis of breast cancer.

```
%sql

select description, count(*) tally from conditions where code = '254837009' group by description
```

```
%sql
create or replace temporary view patient_breast_cancer as
with 
retro_numbered_encounters as (
   SELECT *,
         ROW_NUMBER() OVER (PARTITION BY patient ORDER BY date(start) DESC) AS row_number
   FROM encounters
),
most_recent_encounter as (
  select * from retro_numbered_encounters where row_number = 1
),
breast_ca_conditions as (
  select * from conditions c where c.code = 254837009  -- 'Malignant neoplasm of breast (disorder)'
)
select concat_ws(' ', p.first, p.last) patient_name, p.gender, p.race, p.ethnicity,
        floor (datediff(date(e.start), date(p.birthdate))/365.24) age,
        case when c.code is null then 0 else 1 end as breast_cancer
  from most_recent_encounter e 
    join patients p on e.patient = p.id
    left outer join breast_ca_conditions c on c.patient = e.patient;



select * from patient_breast_cancer limit 5;
```

```
%python

pbc = spark.sql('select * from patient_breast_cancer').toPandas()
# type(pbc.age[0]) # decimal.Decimal
pbc['age'] = pbc['age'].astype(float)
```

```
%python

pbc['age'].dtypes
```

```
%python

import pandas as pd
from sklearn.model_selection import train_test_split
from interpret.glassbox import ExplainableBoostingClassifier

X = pbc[['gender', 'race', 'ethnicity', 'age']]
y = pbc['breast_cancer']

seed = 1
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=seed)

ebm = ExplainableBoostingClassifier(random_state=seed)
ebm.fit(X_train, y_train)
```

```
%python

from interpret import show

ebm_global = ebm.explain_global()
show(ebm_global)
```

```
%python

ebm_local = ebm.explain_local(X_test[10:15], y_test[10:15])
show(ebm_local)
```

```
%python
# y_test

p_test = ebm.predict_proba(X_test)[:,1]

actual_predicted_pdf = pd.DataFrame({'actual':y_test, 'predicted_probability':p_test})

## I'll plot these densities in R. Export the data to the database:
spark.createDataFrame(actual_predicted_pdf).createOrReplaceTempView("actual_predicted")

## or make a permanent table:
# actual_predicted_pdf.write.mode("overwrite").saveAsTable("actual_predicted")
```

```
%python

from interpret import perf
roc = perf.ROC(ebm.predict_proba, feature_names=X_train.columns)

roc_explanation = roc.explain_perf(X_test, y_test)
show(roc_explanation)
```
